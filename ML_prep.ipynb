## Modeling using linear regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
data = pd.read_csv('modelling_table.csv')

# Define features and separate targets
X = data.drop(columns=['Team 1 Score', 'Team 2 Score'])

# Target variables
y_team1 = data['Team 1 Score']
y_team2 = data['Team 2 Score']

# Splitting data for Team 1
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y_team1, test_size=0.2, random_state=42)

# Splitting data for Team 2
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y_team2, test_size=0.2, random_state=42)

# Model for Team 1 Score
model_team1 = LinearRegression()
model_team1.fit(X_train_1, y_train_1)
y_pred_1 = model_team1.predict(X_test_1)
mse_team1 = mean_squared_error(y_test_1, y_pred_1)
r2_team1 = r2_score(y_test_1, y_pred_1)

# Model for Team 2 Score
model_team2 = LinearRegression()
model_team2.fit(X_train_2, y_train_2)
y_pred_2 = model_team2.predict(X_test_2)
mse_team2 = mean_squared_error(y_test_2, y_pred_2)
r2_team2 = r2_score(y_test_2, y_pred_2)

# Print the results for both models
print("Team 1 Score Prediction - MSE:", mse_team1, "R^2:", r2_team1)
print("Team 2 Score Prediction - MSE:", mse_team2, "R^2:", r2_team2)
```
