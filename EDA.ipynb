```python
import pandas as pd
# Remove the limit on the number of columns displayed
pd.set_option('display.max_columns', None)
```

```python
table_df = pd.read_csv("finaltable1.csv")
```

```python
table_df.columns
```

```python
table_df.info()
```

```python
# Selecting relevant numeric columns for the correlation heatmap
relevant_columns = [
    'Home Score', 'Away Score', 'Home Goals Scored', 'Away Goals Scored', 
    'Home Goal Difference', 'Away Goal Difference', 'Home Points', 'Away Points', 
    'xG', 'xG.1', 'home_xg', 'away_xg', 'home_np_xg', 'away_np_xg', 
    'home_np_xg_difference', 'away_np_xg_difference', 'home_ppda', 'away_ppda', 
    'home_deep_completions', 'away_deep_completions', 'Temperature', 'Attendance', 'home_team_fdr','away_team_fdr','B365H',
    'B365D','B365A'
]

# Subset the DataFrame with the relevant columns
subset_df = table_df[relevant_columns]

# Plot the correlation heatmap for the selected columns
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))
sns.heatmap(subset_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap for Relevant Columns")
plt.show()
```
