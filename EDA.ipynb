```python
import numpy as np
import pandas as pd
# Remove the limit on the number of columns displayed
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("modelling_table.csv")
```

```python
df.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the original data
data = df

# Select only numeric features for correlation calculation
numeric_features = data.select_dtypes(include=[np.number])

# Calculate correlation with 'Team 1 Score' and 'Team 2 Score'
score_correlations_1 = numeric_features.corrwith(data['Team 1 Score']).sort_values(ascending=True)
score_correlations_2 = numeric_features.corrwith(data['Team 2 Score']).sort_values(ascending=True)

# Plotting the correlation values for Team 1 Score
plt.figure(figsize=(10, 20))
score_correlations_1.plot(kind='barh', color='skyblue')
plt.title('Feature Correlation with Team 1 Score')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()

# Plotting the correlation values for Team 2 Score
plt.figure(figsize=(10, 20))
score_correlations_2.plot(kind='barh', color='lightgreen')
plt.title('Feature Correlation with Team 2 Score')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data_path = 'modelling_table.csv'  # Update your path accordingly
data = pd.read_csv(data_path)

# Example features you might want to plot
features_to_plot = ['Prev Team 1 Form', 'Avg xG Team 1', 'Avg PPDA Team 1']

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
for idx, feature in enumerate(features_to_plot):
    sns.scatterplot(x=feature, y='Team 1 Score', data=data, ax=axes[idx])
    axes[idx].set_title(f'Scatter plot: {feature} vs Team 1 Score')
    axes[idx].set_xlabel(feature)
    axes[idx].set_ylabel('Team 1 Score')

plt.tight_layout()
plt.show()
```
