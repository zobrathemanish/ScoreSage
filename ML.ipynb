# Data Preprocessing

#### Drop first 10 rows

#### Change data types

```python
import pandas as pd

# Load your data
# df = pd.read_csv('your_file.csv')

# Convert to integer
df['StadiumID'] = df['StadiumID'].astype(int)

# Convert to category
df['Venue'] = df['Venue'].astype('category')
df['Referee'] = df['Referee'].astype('category')
df['team_2_code'] = df['team_2_code'].astype('category')
df['team_1_code'] = df['team_1_code'].astype('category')

# Convert to float32
df['xG'] = df['xG'].astype('float32')
df['Attendance'] = df['Attendance'].astype('float32')


# View changes
print(df.info())
```

#### Label encode categorical columns and scaling some numerical columns

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from scipy import sparse

# Define categorical and numerical features
categorical_features = ['Venue', 'Referee', 'team_2_code', 'team_1_code']
numerical_features = ['xG', 'xG.1', 'Attendance', 'Latitude', 'Longitude']

# Identify other columns to retain (not part of preprocessing)
other_columns = [col for col in df.columns if col not in categorical_features + numerical_features]

# One-hot encoding and scaling
one_hot = OneHotEncoder()
scaler = StandardScaler()

# Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', scaler, numerical_features),
        ('cat', one_hot, categorical_features)
    ])

# Fit and transform data
data_preprocessed = preprocessor.fit_transform(df)

# Check if the output is a sparse matrix and convert to dense array if true
if sparse.issparse(data_preprocessed):
    data_preprocessed = data_preprocessed.toarray()

# Generate column names for the transformed features
num_features_transformed = numerical_features
cat_features_transformed = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
all_features = num_features_transformed + list(cat_features_transformed)

# Create the transformed DataFrame
data_preprocessed_df = pd.DataFrame(data_preprocessed, columns=all_features)

# Add back the other columns that were not processed
for col in other_columns:
    data_preprocessed_df[col] = df[col].values

# Check the updated DataFrame
data_preprocessed_df.head()
```

#### Transformations for skewness

```python
import numpy as np
from scipy.stats import boxcox
from scipy.special import boxcox1p

# Load your data if not already loaded
data = data_preprocessed_df

# Identify highly skewed features (skewness > 1 or skewness < -1)
highly_skewed = data.skew()
highly_skewed = highly_skewed[(highly_skewed > 1) | (highly_skewed < -1)]

# Transformations
for feature in highly_skewed.index:
    if feature in ['H2H_Team_1_Wins_Last_4', 'H2H_Team_2_Wins_Last_4', 'H2H_Draws_Last_4', 'team2_ppda', 'team1_ppda', 
                   'B365H', 'B365D', 'B365A', 'Cum Team 1 Goals Scored', 'Cum Team 2 Goals Scored', 
                   'Cum Team 1 Goals Conceded', 'team2_np_xg']:
        # Apply log1p transformation for positive skewness
        data[feature] = np.log1p(data[feature])
    elif feature in ['Avg PPDA Team 1']:
        # For negative skewness, if any feature is negatively skewed (for illustration)
        # Adjust with squaring if it's negative skew, though your data might not need it.
        data[feature] = data[feature]**2
    else:
        # Ensure all data are positive before applying Box-Cox
        # Shift the data by adding the absolute value of the smallest value plus one
        min_val = data[feature].min()
        shift = 1 - min_val if min_val <= 0 else 0
        data[feature], fitted_lambda = boxcox(data[feature] + shift)

# Check skewness after transformations
new_skewness = data[highly_skewed.index].skew()
print("New skewness of the features:")
print(new_skewness)
```

#### PCA

Standaradize the data: PCA requires features to be on the same scale because it emphasizes variance.

```python
data_preprocessed_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assuming 'data_preprocessed_df' is the DataFrame with actual data
features = data_preprocessed_df.drop(['Team 1 Score', 'Team 2 Score'], axis=1)  # Exclude the target variables for PCA
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)  # Standardize the features

# Apply PCA to reduce dimensionality while retaining 95% of the variance
pca = PCA(n_components=0.95)
principal_components = pca.fit_transform(features_scaled)
print("Number of PCA components:", pca.n_components_)

# If you want to use these components for further analysis or model building:
pca_features_df = pd.DataFrame(principal_components)
pca_features_df['Team 1 Score'] = data_preprocessed_df['Team 1 Score']
pca_features_df['Team 2 Score'] = data_preprocessed_df['Team 2 Score']

# Now you can use pca_features_df for your modeling tasks
```

# Train using linear regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# For Team 1 Score Prediction
X = pca_features_df.drop(['Team 1 Score', 'Team 2 Score'], axis=1)
y_team1 = pca_features_df['Team 1 Score']

# Split the dataset for Team 1 Score
X_train, X_test, y_train_team1, y_test_team1 = train_test_split(X, y_team1, test_size=0.2, random_state=42)

# Train a linear regression model
model_team1 = LinearRegression()
model_team1.fit(X_train, y_train_team1)

# Make predictions for Team 1 Score
y_pred_team1 = model_team1.predict(X_test)

# Evaluate the model for Team 1 Score
mse_team1 = mean_squared_error(y_test_team1, y_pred_team1)
r2_team1 = r2_score(y_test_team1, y_pred_team1)

print("Team 1 Score Prediction:")
print(f"Mean Squared Error: {mse_team1}")
print(f"R-squared: {r2_team1}")

# For Team 2 Score Prediction
y_team2 = pca_features_df['Team 2 Score']

# Split the dataset for Team 2 Score
X_train, X_test, y_train_team2, y_test_team2 = train_test_split(X, y_team2, test_size=0.2, random_state=42)

# Train a linear regression model
model_team2 = LinearRegression()
model_team2.fit(X_train, y_train_team2)

# Make predictions for Team 2 Score
y_pred_team2 = model_team2.predict(X_test)

# Evaluate the model for Team 2 Score
mse_team2 = mean_squared_error(y_test_team2, y_pred_team2)
r2_team2 = r2_score(y_test_team2, y_pred_team2)

print("\nTeam 2 Score Prediction:")
print(f"Mean Squared Error: {mse_team2}")
print(f"R-squared: {r2_team2}")
```

# Classification

```python
# Create a target column for classification: 1 (Win), 0 (Draw), -1 (Lose)
pca_features_df['Match Outcome'] = (pca_features_df['Team 1 Score'] > pca_features_df['Team 2 Score']).astype(int) - \
                                   (pca_features_df['Team 1 Score'] < pca_features_df['Team 2 Score']).astype(int)
```

#### Train a classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define features and target
X_class = pca_features_df.drop(['Team 1 Score', 'Team 2 Score', 'Match Outcome'], axis=1)
y_class = pca_features_df['Match Outcome']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Train a Random Forest Classifier
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)

# Make predictions
y_pred_class = classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_class)
print("Accuracy for Match Outcome Classification:", accuracy)
```
