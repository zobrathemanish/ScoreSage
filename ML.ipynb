# Data Preprocessing

#### Drop first 10 rows

```python
dataset = dataset.drop(dataset.index[:10])
df = dataset
```

#### Change data types

```python
import pandas as pd


# Convert to integer
df['StadiumID'] = df['StadiumID'].astype(int)

# Convert to category
df['Venue'] = df['Venue'].astype('category')
df['Referee'] = df['Referee'].astype('category')
df['team_2_code'] = df['team_2_code'].astype('category')
df['team_1_code'] = df['team_1_code'].astype('category')

# Convert to float32
df['Attendance'] = df['Attendance'].astype('float32')


# View changes
print(df.info())
```

```python
df.to_csv("modelling_table.csv")
```

```python
df.columns
```

## Modeling

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

data = df

# Remove unnecessary index columns and non-numeric columns
data_cleaned = data.drop(columns=[data.columns[0]])  # Assuming the first column is an index column
features = data_cleaned.select_dtypes(include=[np.number]).drop(columns=['Team 1 Score', 'Team 2 Score'])
targets = data_cleaned[['Team 1 Score', 'Team 2 Score']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95)  # Retain 95% of variance
features_reduced = pca.fit_transform(features_scaled)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features_reduced, targets, test_size=0.2, random_state=42)

# Initialize and train the regression models
lr_team1 = LinearRegression()
lr_team2 = LinearRegression()

lr_team1.fit(X_train, y_train['Team 1 Score'])
lr_team2.fit(X_train, y_train['Team 2 Score'])

# Predict the scores
predictions_team1 = lr_team1.predict(X_test)
predictions_team2 = lr_team2.predict(X_test)

# Calculate the mean squared errors
mse_team1 = mean_squared_error(y_test['Team 1 Score'], predictions_team1)
mse_team2 = mean_squared_error(y_test['Team 2 Score'], predictions_team2)

# Calculate the R^2 scores
r2_team1 = r2_score(y_test['Team 1 Score'], predictions_team1)
r2_team2 = r2_score(y_test['Team 2 Score'], predictions_team2)

# Print results
print(f"Mean Squared Error for Team 1: {mse_team1:.4f}")
print(f"Mean Squared Error for Team 2: {mse_team2:.4f}")
print(f"R^2 Score for Team 1: {r2_team1:.4f}")
print(f"R^2 Score for Team 2: {r2_team2:.4f}")
```
