## Weather data integration

```python
# Combine Date and Time into a single column for API query
merged_df['Datetime'] = merged_df['Date'] + 'T' + merged_df['Time'] + ':00'
```

```python
import openmeteo_requests
import requests_cache
import pandas as pd
from datetime import datetime, timedelta
from retry_requests import retry

# Setup caching and retry mechanism
cache_session = requests_cache.CachedSession('.cache', expire_after=-1)
retry_session = retry(cache_session, retries=5, backoff_factor=0.2)
openmeteo = openmeteo_requests.Client(session=retry_session)

# Function to round time to the nearest hour
def round_to_nearest_hour(date_str, time_str):
    dt = datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
    rounded_dt = dt.replace(minute=0) + timedelta(hours=1 if dt.minute >= 30 else 0)
    return pd.to_datetime(rounded_dt, utc=True)  # Ensure timezone-aware datetime

# Function to fetch weather data
def fetch_weather_data(latitude, longitude, date):
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": date,
        "end_date": date,
        "hourly": "temperature_2m,precipitation"
    }
    try:
        responses = openmeteo.weather_api(url, params=params)
        response = responses[0]
        return response.Hourly()
    except Exception as e:
        print(f"Failed to fetch data for {latitude}, {longitude} on {date}: {e}")
        return None

# Add weather data to merged_df
weather_data_list = []
for _, row in merged_df.iterrows():
    date = row['Date']
    time = row['Time']
    latitude = row['Latitude']
    longitude = row['Longitude']
    
    # Round time to the nearest hour
    rounded_datetime = round_to_nearest_hour(date, time)
    
    # Fetch weather data for valid dates
    if date > '2024-12-18':  # Skip unsupported dates
        print(f"Skipping unsupported date: {date}")
        weather_data_list.append({'Temperature': None, 'Precipitation': None})
        continue

    weather = fetch_weather_data(latitude, longitude, date)
    if weather:
        # Extract data from response
        hourly_temperature = weather.Variables(0).ValuesAsNumpy()
        hourly_precipitation = weather.Variables(1).ValuesAsNumpy()
        hourly_times = pd.date_range(
            start=pd.to_datetime(weather.Time(), unit="s", utc=True),
            end=pd.to_datetime(weather.TimeEnd(), unit="s", utc=True),
            freq=pd.Timedelta(seconds=weather.Interval()),
            inclusive="left"
        )
        
        # Create a DataFrame for hourly data
        hourly_df = pd.DataFrame({
            "Datetime": hourly_times,
            "Temperature_2m": hourly_temperature,
            "Precipitation": hourly_precipitation
        })

        # Match the rounded datetime in the hourly data
        matched_row = hourly_df.loc[hourly_df['Datetime'] == rounded_datetime]
        if not matched_row.empty:
            temperature = matched_row['Temperature_2m'].values[0]
            precipitation = matched_row['Precipitation'].values[0]
            weather_data_list.append({'Temperature': temperature, 'Precipitation': precipitation})
        else:
            print(f"Rounded Datetime {rounded_datetime} not found in hourly data")
            weather_data_list.append({'Temperature': None, 'Precipitation': None})
    else:
        weather_data_list.append({'Temperature': None, 'Precipitation': None})

# Add weather data to merged_df
weather_df = pd.DataFrame(weather_data_list)
merged_df = pd.concat([merged_df, weather_df], axis=1)

# Display the final DataFrame
print(merged_df)

# Optionally save to CSV
# merged_df.to_csv("merged_with_weather.csv", index=False)
```

Consistent Timezone Awareness:

The round_to_nearest_hour function now returns a timezone-aware datetime object (utc=True).
The Hourly Time index is already in UTC.


Datetime Conversion:

The Rounded Datetime is compared directly with the Datetime column in hourly_df as datetime objects, ensuring consistency.


Improved Debugging:

Retains debugging messages for unmatched datetimes.

```python
def add_h2h_results_to_df(data):

    # Ensure 'Date' column is in datetime format
    data['Date'] = pd.to_datetime(data['Date'])

    # Initialize new columns
    data['H2H_Home_Wins_Last_4'] = 0
    data['H2H_Away_Wins_Last_4'] = 0
    data['H2H_Draws_Last_4'] = 0

    # Iterate through rows in the original dataset (excluding historical rows for integration)
    for index, row in data.iterrows():
        home_team = row['Home Team']
        away_team = row['Away Team']
        match_date = row['Date']

        # Filter H2H matches up to (but not including) the current match date
        h2h_matches = data[
            (data['Date'] < match_date) &  # Past matches only
            (((data['Home Team'] == home_team) & (data['Away Team'] == away_team)) |
             ((data['Home Team'] == away_team) & (data['Away Team'] == home_team)))
        ]

        # If no matches are found, continue with default values
        if h2h_matches.empty:
            continue

        # Sort by date and take the last 4 matches
        h2h_matches = h2h_matches.sort_values(by='Date', ascending=False).head(4)

        # Calculate outcomes
        home_wins = 0
        away_wins = 0
        draws = 0

        for _, h2h_row in h2h_matches.iterrows():
            if h2h_row['Home Score'] > h2h_row['Away Score']:
                if h2h_row['Home Team'] == home_team:
                    home_wins += 1
                else:
                    away_wins += 1
            elif h2h_row['Home Score'] < h2h_row['Away Score']:
                if h2h_row['Away Team'] == away_team:
                    away_wins += 1
                else:
                    home_wins += 1
            else:
                draws += 1

        # Update the current row with H2H results
        data.at[index, 'H2H_Home_Wins_Last_4'] = home_wins
        data.at[index, 'H2H_Away_Wins_Last_4'] = away_wins
        data.at[index, 'H2H_Draws_Last_4'] = draws

    return data
```

```python
# Compute H2H stats
combined_data = add_h2h_results_to_df(combined_data)

# Display a sample of the updated DataFrame
print(combined_data[['Home Team', 'Away Team', 'Date', 'H2H_Home_Wins_Last_4', 'H2H_Away_Wins_Last_4', 'H2H_Draws_Last_4']].head(10))
```

```python
team_name_mapping = {
    'Man United': 'Manchester Utd',
    'Ipswich': 'Ipswich Town',
    'Arsenal': 'Arsenal',
    'Everton': 'Everton',
    'Newcastle': 'Newcastle Utd',
    "Nott'm Forest": 'Nottingham Forest',
    'West Ham': 'West Ham',
    'Brentford': 'Brentford',
    'Chelsea': 'Chelsea',
    'Leicester': 'Leicester City',
    'Brighton': 'Brighton',
    'Crystal Palace': 'Crystal Palace',
    'Fulham': 'Fulham',
    'Man City': 'Manchester City',
    'Southampton': 'Southampton',
    'Tottenham': 'Tottenham',
    'Aston Villa': 'Aston Villa',
    'Bournemouth': 'Bournemouth',
    'Wolves': 'Wolves',
    'Liverpool': 'Liverpool'
}
# Function to standardize team names in a column
def standardize_team_names(odds_data_df, column, mapping):
    odds_data_df[column] = odds_data_df[column].replace(mapping)
    return odds_data_df

# Apply standardization to both Home Team and Away Team columns
odds_data_df = standardize_team_names(odds_data_df, 'HomeTeam', team_name_mapping)
odds_data_df = standardize_team_names(odds_data_df, 'AwayTeam', team_name_mapping)

# Convert the Date column to datetime format and then format it to YYYY-MM-DD
odds_data_df['Date'] = pd.to_datetime(odds_data_df['Date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')
```

Cumulatatitve stats

```python
# Initialize new columns for cumulative averages
stats = ['HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']
for stat in stats:
    odds_data_df[f'Cumulative_Avg_{stat}'] = 0.0

# Initialize dictionaries to track total stats and number of games for each team
team_stats_totals = {stat: {} for stat in stats}
team_game_counts = {stat: {} for stat in stats}

# Iterate through the DataFrame row by row
for index, row in odds_data_df.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    
    # Calculate cumulative averages for each stat
    for stat in stats:
        home_stat = stat  # Home version of the stat
        away_stat = stat  # Away version of the stat

        # Initialize stats for home and away teams if not already in dictionaries
        if home_team not in team_stats_totals[home_stat]:
            team_stats_totals[home_stat][home_team] = 0
            team_game_counts[home_stat][home_team] = 0
        if away_team not in team_stats_totals[away_stat]:
            team_stats_totals[away_stat][away_team] = 0
            team_game_counts[away_stat][away_team] = 0

        # Assign cumulative average to the new columns
        if team_game_counts[home_stat][home_team] > 0:
            odds_data_df.at[index, f'Cumulative_Avg_{home_stat}'] = (
                team_stats_totals[home_stat][home_team] / team_game_counts[home_stat][home_team]
            )
        if team_game_counts[away_stat][away_team] > 0:
            odds_data_df.at[index, f'Cumulative_Avg_{away_stat}'] = (
                team_stats_totals[away_stat][away_team] / team_game_counts[away_stat][away_team]
            )

        # Update totals and game counts with the current game's stats
        team_stats_totals[home_stat][home_team] += row[home_stat]
        team_game_counts[home_stat][home_team] += 1
        team_stats_totals[away_stat][away_team] += row[away_stat]
        team_game_counts[away_stat][away_team] += 1

# Display the updated DataFrame with cumulative averages
odds_data_df.head(13)
```

```python
# Create a composite key by concatenating Date, HomeTeam, and AwayTeam
odds_data_df['composite_key'] = odds_data_df['Date'] + '_' + odds_data_df['HomeTeam'] + '_' + odds_data_df['AwayTeam']
odds_data_df = odds_data_df.drop('Date', axis=1)
# Verify the result
print(odds_data_df[['HomeTeam', 'AwayTeam', 'composite_key']].head(10))
```
