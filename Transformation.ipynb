## Weather data integration

```python
# Combine Date and Time into a single column for API query
fbref_matches['Datetime'] = fbref_matches['Date'] + 'T' + fbref_matches['Time'] + ':00'
```

```python
import openmeteo_requests
import requests_cache
import pandas as pd
from datetime import datetime, timedelta
from retry_requests import retry

# Setup caching and retry mechanism
cache_session = requests_cache.CachedSession('.cache', expire_after=-1)
retry_session = retry(cache_session, retries=5, backoff_factor=0.2)
openmeteo = openmeteo_requests.Client(session=retry_session)

# Function to round time to the nearest hour
def round_to_nearest_hour(date_str, time_str):
    dt = datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
    rounded_dt = dt.replace(minute=0) + timedelta(hours=1 if dt.minute >= 30 else 0)
    return pd.to_datetime(rounded_dt, utc=True)  # Ensure timezone-aware datetime

# Function to fetch weather data
def fetch_weather_data(latitude, longitude, date):
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": date,
        "end_date": date,
        "hourly": "temperature_2m,precipitation"
    }
    try:
        responses = openmeteo.weather_api(url, params=params)
        response = responses[0]
        return response.Hourly()
    except Exception as e:
        print(f"Failed to fetch data for {latitude}, {longitude} on {date}: {e}")
        return None

# Add weather data to merged_df
weather_data_list = []
for _, row in fbref_matches.iterrows():
    date = row['Date']
    time = row['Time']
    latitude = row['Latitude']
    longitude = row['Longitude']
    
    # Round time to the nearest hour
    rounded_datetime = round_to_nearest_hour(date, time)
    
    # Fetch weather data for valid dates
    if date > '2024-12-18':  # Skip unsupported dates
        print(f"Skipping unsupported date: {date}")
        weather_data_list.append({'Temperature': None, 'Precipitation': None})
        continue

    weather = fetch_weather_data(latitude, longitude, date)
    if weather:
        # Extract data from response
        hourly_temperature = weather.Variables(0).ValuesAsNumpy()
        hourly_precipitation = weather.Variables(1).ValuesAsNumpy()
        hourly_times = pd.date_range(
            start=pd.to_datetime(weather.Time(), unit="s", utc=True),
            end=pd.to_datetime(weather.TimeEnd(), unit="s", utc=True),
            freq=pd.Timedelta(seconds=weather.Interval()),
            inclusive="left"
        )
        
        # Create a DataFrame for hourly data
        hourly_df = pd.DataFrame({
            "Datetime": hourly_times,
            "Temperature_2m": hourly_temperature,
            "Precipitation": hourly_precipitation
        })

        # Match the rounded datetime in the hourly data
        matched_row = hourly_df.loc[hourly_df['Datetime'] == rounded_datetime]
        if not matched_row.empty:
            temperature = matched_row['Temperature_2m'].values[0]
            precipitation = matched_row['Precipitation'].values[0]
            weather_data_list.append({'Temperature': temperature, 'Precipitation': precipitation})
        else:
            print(f"Rounded Datetime {rounded_datetime} not found in hourly data")
            weather_data_list.append({'Temperature': None, 'Precipitation': None})
    else:
        weather_data_list.append({'Temperature': None, 'Precipitation': None})

# Add weather data to merged_df
weather_df = pd.DataFrame(weather_data_list)
fbref_matches = pd.concat([fbref_matches, weather_df], axis=1)

# Display the final DataFrame
print(fbref_matches)

# Optionally save to CSV
# merged_df.to_csv("merged_with_weather.csv", index=False)
```

Consistent Timezone Awareness:

The round_to_nearest_hour function now returns a timezone-aware datetime object (utc=True).
The Hourly Time index is already in UTC.


Datetime Conversion:

The Rounded Datetime is compared directly with the Datetime column in hourly_df as datetime objects, ensuring consistency.


Improved Debugging:

Retains debugging messages for unmatched datetimes.

```python
def add_total_h2h_results_to_df(data):
    # Ensure 'Date' column is in datetime format
    data['Date'] = pd.to_datetime(data['Date'])

    # Initialize new columns
    data['H2H_Home_Total_Wins_Last_4'] = 0
    data['H2H_Away_Total_Wins_Last_4'] = 0
    data['H2H_Draws_Last_4'] = 0

    # Iterate through rows
    for index, row in data.iterrows():
        home_team = row['Home Team']
        away_team = row['Away Team']
        match_date = row['Date']

        # Filter H2H matches up to (but not including) the current match date
        h2h_matches = data[
            (data['Date'] < match_date) &  # Past matches only
            (((data['Home Team'] == home_team) | (data['Away Team'] == home_team)) |
             ((data['Home Team'] == away_team) | (data['Away Team'] == away_team)))
        ]

        # If no matches are found, continue with default values
        if h2h_matches.empty:
            continue

        # Sort by date and take the last 4 matches
        h2h_matches = h2h_matches.sort_values(by='Date', ascending=False).head(4)

        # Calculate outcomes
        home_total_wins = 0
        away_total_wins = 0
        draws = 0

        for _, h2h_row in h2h_matches.iterrows():
            if h2h_row['Home Score'] > h2h_row['Away Score']:
                # If the Home Team in the historical match is the current Home Team
                if h2h_row['Home Team'] == home_team:
                    home_total_wins += 1
                # If the Home Team in the historical match is the current Away Team
                elif h2h_row['Home Team'] == away_team:
                    away_total_wins += 1
            elif h2h_row['Home Score'] < h2h_row['Away Score']:
                # If the Away Team in the historical match is the current Home Team
                if h2h_row['Away Team'] == home_team:
                    home_total_wins += 1
                # If the Away Team in the historical match is the current Away Team
                elif h2h_row['Away Team'] == away_team:
                    away_total_wins += 1
            else:
                # Count draws
                draws += 1

        # Update the current row with the total H2H results
        data.at[index, 'H2H_Home_Total_Wins_Last_4'] = home_total_wins
        data.at[index, 'H2H_Away_Total_Wins_Last_4'] = away_total_wins
        data.at[index, 'H2H_Draws_Last_4'] = draws

    return data
```
